## Pyramids visualization

```python
female_color = "#ee7a87"
male_color = "#4682b4"
#France 1835:
df = pd.read_csv("France.csv", sep=';')

df['Age'] = df['Age'].astype(str)
df['Age'] = df['Age'].replace({'0': '0-4', '1-4': '0-4'})
df = (df.groupby(['Year','Age'], as_index=False)[['Female','Male','Total']].sum())
df['age_start'] = (
    df['Age']
      .str.replace('+','', regex=False)
      .str.split('-', n=1, expand=True)[0]
      .astype(int)
)

df_new = df[df['Year']=='1835'].sort_values('age_start')

population_df = pd.DataFrame({"Age": df_new['Age'], "Male": df_new['Male'], "Female": df_new['Female']})
population_df["Female_Left"] = 0
population_df["Female_Width"] = population_df["Female"]
population_df["Male_Left"] = -population_df["Male"]
population_df["Male_Width"] = population_df["Male"]

fig = plt.figure(figsize=(12,10), dpi = 500)
plt.barh(y=population_df["Age"], width=-population_df["Male_Width"], color=male_color, label="")
plt.barh(y=population_df["Age"], width=population_df["Female_Width"], color=female_color, label="")
plt.text(-max(population_df["Male_Width"]) * 0.8, population_df["Age"].max(), "", 
         fontsize=15, ha='center')
plt.text(max(population_df["Female_Width"]) * 0.8, population_df["Age"].max(), "", 
         fontsize=15, ha='center')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.legend(loc="best");
max_count = max(population_df["Female_Width"].max(), population_df["Male_Width"].max())
plt.xticks(
    ticks=[-max_count, -max_count/2, 0, max_count/2, max_count],
    labels=[f"{abs(int(x)):,}" for x in [-max_count, -max_count/2, 0, max_count/2, max_count]]
)
plt.xlabel("", fontsize=14, fontweight="bold")
plt.ylabel(" ( )", fontsize=14, fontweight="bold")
plt.title("  , 1835", pad=10, fontsize=20, fontweight="bold");
```
